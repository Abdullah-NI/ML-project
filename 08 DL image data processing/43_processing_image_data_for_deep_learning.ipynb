**converting the RGB image to grayscale image using opencv**

(means convert colour image into black and white)

```python
#importing the open cv library
import cv2
```

```python
img=cv2.imread('/content/drive/MyDrive/AllDataset/42 dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg',grayscale_image)
```
