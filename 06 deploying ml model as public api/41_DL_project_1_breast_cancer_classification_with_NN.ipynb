Separating the features and target

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data & Testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sclar=StandardScaler()
x_train_std=sclar.fit_transform(x_train)
x_test_std=sclar.transform(x_test)
```

**Building the Neural Network**

image.png

```python
#importing trnsorflow and keras
import tensorflow as tf
tf.random.set_seed(3)   # ye random state ki trah hai means jab bhi ham code run karege same accuracy aaygi
from tensorflow import keras
```

```python
#setting up the layer of neural network
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),  # input layer with 30 neuroon bcz of 30 feature in x
                        keras.layers.Dense(20,activation='relu'), # hidden layer with 20 neuroon
                       # keras.layers.Dense(13,activation='sigmoid'), # 2nd hidden layer with 13 neuroon
                        keras.layers.Dense(2,activation='sigmoid') # output layer with 2 neuroon bcz we have to classes in target 1 and 0  means 1 --> Benign
                                                                                                                                                 #0 --> Malignant
                        ])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
```

```python
#training the neural network
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

vidualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

**<h1>Note**

**Accuracy**

**Definition**: Accuracy is a metric that measures the percentage of correctly classified instances over the total number of instances in a dataset. It's a common evaluation metric for classification models.
**Calculation**: Accuracy = (Number of Correct Predictions) / (Total Number of Predictions)
**In your code:** history.history['accuracy'] stores the accuracy of your model on the training data during each epoch (a full pass over the training dataset).
**Interpretation**: A high accuracy on the training data means that your model is learning to correctly predict the labels for the samples it has seen during training.

**Validation Accuracy**

**Definition**: Validation accuracy is similar to accuracy, but it's calculated on a separate dataset called the validation set, which is not used during training.
**Purpose**: The primary purpose of validation accuracy is to estimate how well your model will generalize to unseen data, providing an indication of its performance on real-world examples. This helps in detecting overfitting, where the model learns the training data too well but fails to perform well on new data.
**In your code:** history.history['val_accuracy'] stores the validation accuracy during each epoch. It's calculated using the portion of data you set aside with validation_split=0.1 in the model.fit function.
**Interpretation**: A high validation accuracy suggests that your model is likely to perform well on unseen data, and that it is not overfitting the training data.

**In the plot you generated:**

The blue line represents the training accuracy (history.history['accuracy']).

The orange line represents the validation accuracy (history.history['val_accuracy']).

Ideally, you want both training and validation accuracy to be high and close to each other. If training accuracy is much higher than validation accuracy, it indicates overfitting. If both are low, the model is underfitting and not learning effectively.
